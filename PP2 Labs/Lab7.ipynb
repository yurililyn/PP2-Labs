```python
import pygame
import datetime

pygame.init()

WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Rotating Clock Hands")

clock_face = pygame.image.load("images/clock.png")
clock_rect = clock_face.get_rect(center=(WIDTH//2, HEIGHT//2))

sec_hand = pygame.image.load("images/sec_hand.png")
sec_rect = sec_hand.get_rect(center=(WIDTH//2, HEIGHT//2))

min_hand = pygame.image.load("images/min_hand.png")
min_rect = min_hand.get_rect(center=(WIDTH//2, HEIGHT//2))


running = True
while running:
    screen.fill((255, 255, 255))
    screen.blit(clock_face, clock_rect)

    now = datetime.datetime.now()
    seconds = now.second
    minutes = now.minute

    sec_correct_angle = 56
    sec_angle = -seconds * 6 + sec_correct_angle

    min_correct_angle = 52
    min_angle = -minutes * 6 - min_correct_angle  

    rotated_sec = pygame.transform.rotate(sec_hand, sec_angle)
    rotated_sec_rect = rotated_sec.get_rect(center=sec_rect.center)

    rotated_min = pygame.transform.rotate(min_hand, min_angle)
    rotated_min_rect = rotated_min.get_rect(center=min_rect.center)

    screen.blit(rotated_min, rotated_min_rect)
    screen.blit(rotated_sec, rotated_sec_rect)

    pygame.display.update()

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

pygame.quit()
```

```python
import pygame
import sys
import random

pygame.init()

WIDTH, HEIGHT = 600, 400
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Pygame Music Player")

pygame.mixer.init()

WHITE = (255, 255, 255)
playlist = [
    "music/fly_me_to_the_moon.mp3",
    "music/nothing.mp3",
    "music/another_one_bites_the_dust.mp3"
]
current_track = 0

is_playing = False
is_paused = False

def load_music(track):
    pygame.mixer.music.load(playlist[track])

def play_music():
    global is_playing, is_paused
    if not is_playing:
        load_music(current_track)
        pygame.mixer.music.play()
        is_playing = True
        is_paused = False
    elif is_paused:
        pygame.mixer.music.unpause()
        is_paused = False

def pause_music():
    global is_paused
    if is_playing and not is_paused:
        pygame.mixer.music.pause()
        is_paused = True

def stop_music():
    global is_playing, is_paused
    pygame.mixer.music.stop()
    is_playing = False
    is_paused = False

def next_song():
    global current_track, is_playing
    if len(playlist) > 1:
        current_track = (current_track + 1) % len(playlist) # для избежания превышения индекса
        load_music(current_track)
        pygame.mixer.music.play()
        is_playing = True

def prev_song():
    global current_track, is_playing
    if len(playlist) > 1:
        current_track = (current_track - 1) % len(playlist)
        load_music(current_track)
        pygame.mixer.music.play()
        is_playing = True

running = True
while running:
    screen.fill(WHITE)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_p:  # "P" - play
                play_music()
            elif event.key == pygame.K_s:  # "S" - stop
                stop_music()
            elif event.key == pygame.K_SPACE:  # "Space" - pause
                pause_music()
            elif event.key == pygame.K_n:  # "N" - next song
                next_song()
            elif event.key == pygame.K_b:  # "B" - previous song
                prev_song()

    pygame.display.update()

pygame.quit()
sys.exit()
```

```python
import pygame
import sys

pygame.init()

WIDTH, HEIGHT = 600, 400
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Moving Ball")

WHITE = (255, 255, 255)
RED = (255, 0, 0)

BALL_RADIUS = 25
ball_x = WIDTH // 2
ball_y = HEIGHT // 2
BALL_SPEED = 20

running = True
while running:
    screen.fill(WHITE)
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP and ball_y - BALL_RADIUS - BALL_SPEED >= 0:
                ball_y -= BALL_SPEED
            elif event.key == pygame.K_DOWN and ball_y + BALL_RADIUS + BALL_SPEED <= HEIGHT:
                ball_y += BALL_SPEED
            elif event.key == pygame.K_LEFT and ball_x - BALL_RADIUS - BALL_SPEED >= 0:
                ball_x -= BALL_SPEED
            elif event.key == pygame.K_RIGHT and ball_x + BALL_RADIUS + BALL_SPEED <= WIDTH:
                ball_x += BALL_SPEED
    
    pygame.draw.circle(screen, RED, (ball_x, ball_y), BALL_RADIUS)
    
    pygame.display.update()

pygame.quit()
sys.exit()
```
