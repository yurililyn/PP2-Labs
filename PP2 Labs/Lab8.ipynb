```python
import pygame
import sys
import random
import os
from pygame.locals import *

# Инициализация Pygame
pygame.init()

# Настройки экрана и кадров
SCREEN_WIDTH = 400
SCREEN_HEIGHT = 600
FPS = 60
SPEED = 7
SCORE = 0
COINS = 0

# Параметры дороги и монеты
ROAD_LEFT = 70
ROAD_RIGHT = 290
COIN_WIDTH = 40

# Таймер и шрифты
FramePerSec = pygame.time.Clock()
font = pygame.font.SysFont("Verdana", 20)
font_small = pygame.font.SysFont("Verdana", 20)

# Цвета
RED = (255, 0, 0)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)

# Загрузка фона
background = pygame.image.load(os.path.join("images", "AnimatedStreet.png"))

# Создание окна
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
screen.fill(WHITE)
pygame.display.set_caption("Racer")

# Класс игрока
class Player(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.image.load(os.path.join("images", "Player.png"))
        self.rect = self.image.get_rect(center=(160, 520))
        self.speed = 7

    def move(self):
        keys = pygame.key.get_pressed()
        if keys[K_LEFT] and self.rect.left > 0:
            self.rect.x -= self.speed
        if keys[K_RIGHT] and self.rect.right < SCREEN_WIDTH:
            self.rect.x += self.speed
        if keys[K_UP] and self.rect.top > 0:
            self.rect.y -= self.speed
        if keys[K_DOWN] and self.rect.bottom < SCREEN_HEIGHT:
            self.rect.y += self.speed

# Класс врага
class Enemy(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.image.load(os.path.join("images", "Enemy.png"))
        self.rect = self.image.get_rect()
        self.rect.centerx = random.randint(40, SCREEN_WIDTH - 40)
        self.rect.top = -self.rect.height
        self.speed = SPEED + 2

    def move(self):
        global SCORE
        self.rect.y += self.speed
        if self.rect.top > SCREEN_HEIGHT:
            SCORE += 1
            self.speed += 0.2
            self.rect.top = -self.rect.height
            self.rect.centerx = random.randint(40, SCREEN_WIDTH - 40)

# Класс монеты
class Coin(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.transform.scale(pygame.image.load(os.path.join("images", "Coin.png")), (COIN_WIDTH, COIN_WIDTH))
        self.respawn()

    def respawn(self):
        x = random.randint(ROAD_LEFT, ROAD_RIGHT)
        y = -COIN_WIDTH
        self.rect = self.image.get_rect(topleft=(x, y))

    def move(self):
        global COINS
        self.rect.y += SPEED

        if pygame.sprite.collide_rect(self, P1):
            COINS += 1
            self.respawn()
        elif self.rect.top > SCREEN_HEIGHT:
            self.respawn()

# Экран окончания игры
def show_game_over():
    screen.fill(BLACK)
    game_over_text = font.render("Game Over!", True, WHITE)
    final_score = font.render(f"Score: {SCORE}", True, WHITE)
    coin_score = font.render(f"Coins collected: {COINS}", True, WHITE)
    howToExitText = font.render("Press ESC to exit", True, WHITE)
    screen.blit(game_over_text, (120, 200))
    screen.blit(final_score, (120, 230))
    screen.blit(coin_score, (120, 260))
    screen.blit(howToExitText, (120, 310))


    pygame.display.update()
    while True:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit(); sys.exit()
            elif event.type == KEYDOWN:
                if event.key == K_ESCAPE:
                    pygame.quit(); sys.exit()

# Сброс игры
def restart_game():
    global SCORE, COINS
    SCORE = 0
    COINS = 0
    enemies.empty()
    coinss.empty()
    all_sprites.empty()

    new_player = Player()
    new_enemy = Enemy()
    new_coin = Coin()

    enemies.add(new_enemy)
    coinss.add(new_coin)
    all_sprites.add(new_player, new_enemy, new_coin)
    return new_player, new_enemy, new_coin

# Инициализация объектов
P1 = Player()
E1 = Enemy()
C1 = Coin()

# Группы спрайтов
enemies = pygame.sprite.Group(E1)
coinss = pygame.sprite.Group(C1)
all_sprites = pygame.sprite.Group(P1, E1, C1)

# Фон
background_y = 0

# Основной игровой цикл
while True:
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit(); sys.exit()

    if SPEED < 15:
        SPEED += 0.001 # Ускорение с каждым новым кадром
    
    # Проверка столкновения с врагом
    if pygame.sprite.spritecollideany(P1, enemies):
        if not show_game_over():
            pygame.quit(); sys.exit()
        P1, E1, C1 = restart_game()
        continue

    # Прокрутка фона
    background_y = (background_y + SPEED) % background.get_height()
    screen.blit(background, (0, background_y))
    screen.blit(background, (0, background_y - background.get_height()))

    # Отображение счёта
    screen.blit(font_small.render(str(SCORE), True, BLACK), (10, 10))
    screen.blit(font_small.render(str(COINS), True, BLACK), (370, 10))

    # Обработка и отрисовка спрайтов
    for entity in all_sprites:
        if isinstance(entity, Player):
            entity.move()
        elif isinstance(entity, Enemy):
            entity.move()
        elif isinstance(entity, Coin):
            entity.move()
        screen.blit(entity.image, entity.rect)

    pygame.display.update()
    FramePerSec.tick(FPS)
```

```python
import pygame
import random
import time

# Инициализация Pygame
pygame.init()

# Константы
WIDTH, HEIGHT = 600, 400
GRID_SIZE = 20
GRID_WIDTH = WIDTH // GRID_SIZE
GRID_HEIGHT = HEIGHT // GRID_SIZE
INITIAL_FPS = 10

# Цвета
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
GREEN = (0, 255, 0)
RED = (255, 0, 0)
BLUE = (0, 0, 255)

# Направления
UP = (0, -1)
DOWN = (0, 1)
LEFT = (-1, 0)
RIGHT = (1, 0)

# Класс змейки
class Snake:
    def __init__(self):
        self.positions = [(GRID_WIDTH // 2, GRID_HEIGHT // 2)]
        self.direction = RIGHT
        self.length = 1
        self.score = 0
        self.level = 1
    
    # Позиция головы
    def get_head_position(self):
        return self.positions[0]
    
    # Обновление позиции головы
    def update(self):
        head = self.get_head_position()
        x, y = self.direction
        new_x = head[0] + x
        new_y = head[1] + y
        
        # Проверка на столкновение со стенами
        if new_x < 0 or new_x >= GRID_WIDTH or new_y < 0 or new_y >= GRID_HEIGHT:
            return True
        
        new_position = (new_x, new_y)
        
        # Проверка на столкновении змейки
        if new_position in self.positions[1:]:
            return True
        
        self.positions.insert(0, new_position)
        if len(self.positions) > self.length:
            self.positions.pop()
            
        return False
    
    # Сброс данных
    def reset(self):
        self.positions = [(GRID_WIDTH // 2, GRID_HEIGHT // 2)]
        self.direction = RIGHT
        self.length = 1
        self.score = 0
        self.level = 1

    # Вывод змейки
    def render(self, surface):
        for position in self.positions:
            rect = pygame.Rect(position[0] * GRID_SIZE, position[1] * GRID_SIZE, GRID_SIZE, GRID_SIZE)
            pygame.draw.rect(surface, GREEN, rect)
            pygame.draw.rect(surface, BLACK, rect, 1)

# Класс еды
class Food:
    def __init__(self, snake_positions):
        self.position = (0, 0)
        self.randomize_position(snake_positions)
    
    # Рандомное появление
    def randomize_position(self, snake_positions):
        while True:
            self.position = (
                random.randint(0, GRID_WIDTH - 1),
                random.randint(0, GRID_HEIGHT - 1)
            )
            if self.position not in snake_positions:
                break
    
    # Вывод еды
    def render(self, surface):
        rect = pygame.Rect(self.position[0] * GRID_SIZE, self.position[1] * GRID_SIZE, GRID_SIZE, GRID_SIZE)
        pygame.draw.rect(surface, RED, rect)
        pygame.draw.rect(surface, BLACK, rect, 1)

# Игровая поверхность
def draw_grid(surface):
    for y in range(0, HEIGHT, GRID_SIZE):
        for x in range(0, WIDTH, GRID_SIZE):
            rect = pygame.Rect(x, y, GRID_SIZE, GRID_SIZE)
            pygame.draw.rect(surface, BLACK, rect, 1)

# Экран окончания игры
def show_game_over(surface, score, level):
    surface.fill(BLACK)
    font = pygame.font.SysFont('arial', 36)
    text = font.render(f"Game Over! Score: {score} Level: {level}", True, WHITE)
    text_rect = text.get_rect(center=(WIDTH//2, HEIGHT//2))
    
    font_small = pygame.font.SysFont('arial', 24)
    exit_text = font_small.render("Press ESC to exit", True, WHITE)
    exit_rect = exit_text.get_rect(center=(WIDTH//2, HEIGHT//2 + 50))
    
    surface.blit(text, text_rect)
    surface.blit(exit_text, exit_rect)
    pygame.display.update()
    
    # Нажатие на ESC - выход
    waiting = True
    while waiting:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                return
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_ESCAPE:
                    waiting = False
                    pygame.quit()
                    return

# Показ текущего счета и уровня
def show_score(surface, score, level):
    font = pygame.font.SysFont('arial', 20)
    score_text = font.render(f"Score: {score}  Level: {level}", True, WHITE)
    surface.blit(score_text, (5, 5))

# Основная функция
def main():
    # Настройка окна, таймер
    screen = pygame.display.set_mode((WIDTH, HEIGHT))
    pygame.display.set_caption("Snake Game")
    clock = pygame.time.Clock()
    
    snake = Snake()
    food = Food(snake.positions)
    
    # Флажки
    running = True
    game_over = False
    current_fps = INITIAL_FPS
    
    # Основной цикл
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            elif event.type == pygame.KEYDOWN:
                if not game_over:
                    # Изменение направления без возможности повернуться в противоположную сторону
                    if event.key == pygame.K_UP and snake.direction != DOWN:
                        snake.direction = UP
                    elif event.key == pygame.K_DOWN and snake.direction != UP:
                        snake.direction = DOWN
                    elif event.key == pygame.K_LEFT and snake.direction != RIGHT:
                        snake.direction = LEFT
                    elif event.key == pygame.K_RIGHT and snake.direction != LEFT:
                        snake.direction = RIGHT
        
        # Обновление змейки
        if not game_over:
            game_over = snake.update()
            
            # Проверка если еда съедена
            if snake.get_head_position() == food.position:
                snake.length += 1
                snake.score += 1
                
                # Каждые 3 еды плюс 1 уровень
                if snake.score % 3 == 0:
                    snake.level += 1
                    # Увеличение скорости каждый уровень (максимальная скорость 20)
                    current_fps = min(20, current_fps + 2)
                
                food.randomize_position(snake.positions)
            
            # Рендеринг
            screen.fill(BLACK)
            draw_grid(screen)
            snake.render(screen)
            food.render(screen)
            show_score(screen, snake.score, snake.level)
            pygame.display.update()
            
            clock.tick(current_fps)
        else:
            show_game_over(screen, snake.score, snake.level)
            running = False
    
    pygame.quit()

# Инициализация main для избежания использования функций в других файлах
if __name__ == "__main__":
    main()
```

```python
import pygame

# Освновная функция
def main():
    # Инициализация Pygame
    pygame.init()

    # Настройки окна и таймер
    screen = pygame.display.set_mode((800, 600))
    clock = pygame.time.Clock()
    
    # Флажки и начальные значения
    radius = 5
    mode = 'draw'  # 'draw', 'rectangle', 'circle', 'eraser'
    color_mode = 'white'
    drawing = False
    start_pos = None
    
    # Лист всех рисунков
    drawings = []
    current_shape = None
    
    # Словарь цветов
    colors = {
        'red': (255, 0, 0),
        'green': (0, 255, 0),
        'blue': (0, 0, 255),
        'yellow': (255, 255, 0),
        'cyan': (0, 255, 255),
        'pink': (255, 0, 255),
        'white': (255, 255, 255),
        'black': (0, 0, 0)
    }
    
    # Шрифт
    font = pygame.font.SysFont('Arial', 16)
    
    # Основной цикл
    while True:
        pressed = pygame.key.get_pressed()
        alt_held = pressed[pygame.K_LALT] or pressed[pygame.K_RALT]
        ctrl_held = pressed[pygame.K_LCTRL] or pressed[pygame.K_RCTRL]
        
        # Изменение размера кисти стрелками вниз и вверх
        if pressed[pygame.K_UP]:
            radius = min(50, radius + 1)
            pygame.time.delay(100)  # Задержка для избежания слишком быстрой смены значений при зажатии
        if pressed[pygame.K_DOWN]:
            radius = max(1, radius - 1)
            pygame.time.delay(100)
        
        # Выход из цикла при нажатии на ESC
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                return
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_F4 and alt_held:
                    return
                if event.key == pygame.K_ESCAPE:
                    return
                
                # Выбор режима
                if event.key == pygame.K_d:
                    mode = 'draw'
                elif event.key == pygame.K_r:
                    mode = 'rectangle'
                elif event.key == pygame.K_c:
                    mode = 'circle'
                elif event.key == pygame.K_e:
                    mode = 'eraser'
                
                # Выбор цвета 1-9
                if event.key == pygame.K_1:
                    color_mode = 'red'
                elif event.key == pygame.K_2:
                    color_mode = 'green'
                elif event.key == pygame.K_3:
                    color_mode = 'blue'
                elif event.key == pygame.K_4:
                    color_mode = 'yellow'
                elif event.key == pygame.K_5:
                    color_mode = 'cyan'
                elif event.key == pygame.K_6:
                    color_mode = 'pink'
                elif event.key == pygame.K_7:
                    color_mode = 'white'
                elif event.key == pygame.K_8:
                    color_mode = 'black'
                
                # Очистка экрана, листа рисунков на delete
                if event.key == pygame.K_DELETE:
                    drawings = []
            
            if event.type == pygame.MOUSEBUTTONDOWN:
                if event.button == 1:
                    drawing = True
                    start_pos = event.pos # Начальная точка

                    # Учитывание всех параметров, режимов
                    current_shape = {
                        'type': mode,
                        'color': colors[color_mode] if mode != 'eraser' else (0, 0, 0),
                        'radius': radius,
                        'points': [event.pos] if mode in ['draw', 'eraser'] else None,
                        'start_pos': event.pos,
                        'end_pos': event.pos
                    }

            # Обработка события отпускания кнопки мыши
            if event.type == pygame.MOUSEBUTTONUP:
                if event.button == 1 and drawing and current_shape:
                    drawing = False

                    # Если прямоугольник или круг
                    if mode in ['rectangle', 'circle']:
                        # То фиксируем конечную позицию
                        current_shape['end_pos'] = event.pos

                    # Для режимов рисования и ластика - достаточно ли точек нарисовано(>1)
                    if mode in ['draw', 'eraser'] and len(current_shape['points']) > 1:

                        # Если да то добавляем в лист рисунков
                        drawings.append(current_shape)
                        
                    # Для прямоугольников и кругов всегда добавляем в лист рисунков
                    elif mode in ['rectangle', 'circle']:
                        drawings.append(current_shape)
                    
                    # Сброс текущей фигуры
                    current_shape = None
            
            # Сам процесс рисования - движение мыши, зажатый ЛКМ и выбранная фигура
            if event.type == pygame.MOUSEMOTION and drawing and current_shape:
                if mode in ['draw', 'eraser']:
                    current_shape['points'].append(event.pos)
                elif mode in ['rectangle', 'circle']:
                    current_shape['end_pos'] = event.pos
        
        screen.fill((0, 0, 0))
        
        # Вывод всех сохраненных фигур
        for shape in drawings:
            if shape['type'] == 'draw':
                draw_lines(screen, shape['points'], shape['radius'], shape['color'])
            elif shape['type'] == 'rectangle':
                draw_rectangle(screen, shape['start_pos'], shape['end_pos'], shape['color'])
            elif shape['type'] == 'circle':
                draw_circle(screen, shape['start_pos'], shape['end_pos'], shape['color'])
            elif shape['type'] == 'eraser':
                draw_lines(screen, shape['points'], shape['radius'], shape['color'])
        
        # Вывод всех нарисованных фигур
        if drawing and current_shape:
            if mode in ['draw', 'eraser']:
                draw_lines(screen, current_shape['points'], current_shape['radius'], current_shape['color'])
            elif mode == 'rectangle':
                draw_rectangle(screen, current_shape['start_pos'], current_shape['end_pos'], current_shape['color'])
            elif mode == 'circle':
                draw_circle(screen, current_shape['start_pos'], current_shape['end_pos'], current_shape['color'])
        
        # Вывод UI, интерфейса
        draw_ui(screen, font, mode, color_mode, radius)
        
        pygame.display.flip()
        clock.tick(60) # FPS

def draw_lines(screen, points, width, color):
    if len(points) < 2:
        return
    
    for i in range(len(points) - 1):
        pygame.draw.line(screen, color, points[i], points[i+1], width)

def draw_rectangle(screen, start_pos, end_pos, color):
    rect = pygame.Rect(
        min(start_pos[0], end_pos[0]),
        min(start_pos[1], end_pos[1]),
        abs(end_pos[0] - start_pos[0]),
        abs(end_pos[1] - start_pos[1])
    )
    pygame.draw.rect(screen, color, rect, 2)

def draw_circle(screen, start_pos, end_pos, color):
    radius = int(((end_pos[0] - start_pos[0])**2 + (end_pos[1] - start_pos[1])**2)**0.5)
    pygame.draw.circle(screen, color, start_pos, radius, 2)

def draw_ui(screen, font, mode, color_mode, radius):
    # Получение текущей информации режимов и размера кисти
    tools = [
        f"Tools: D=Draw ({'active' if mode == 'draw' else ''})",
        f"R=Rectangle ({'active' if mode == 'rectangle' else ''})",
        f"C=Circle ({'active' if mode == 'circle' else ''})",
        f"E=Eraser ({'active' if mode == 'eraser' else ''})",
        f"Current: {mode}",
        f"Size: {radius} (UP/DOWN arrows to change)"
    ]
    
    # Текущий цвет
    colors = [
        "Colors:",
        "1=Red",
        "2=Green",
        "3=Blue",
        "4=Yellow",
        "5=Cyan",
        "6=Pink",
        "7=White",
        "8=Black",
        f"Current: {color_mode}"
    ]
    
    # Контроллеры очистки и выхода из программы
    controls = [
        "DELETE=Clear screen",
        "ESC=Quit"
    ]
    
    # Вывод информации о режиме и размере кисти
    y_offset = 10
    for i, text in enumerate(tools):
        text_surface = font.render(text, True, (255, 255, 255))
        screen.blit(text_surface, (10, y_offset + i * 20))
    
    # Вывод информации о текущем цвете
    y_offset = 150
    for i, text in enumerate(colors):
        text_surface = font.render(text, True, (255, 255, 255))
        screen.blit(text_surface, (10, y_offset + i * 20))
    
    # Вывод контроллеров очистки и выхода из програмы
    y_offset = 350
    for i, text in enumerate(controls):
        text_surface = font.render(text, True, (255, 255, 255))
        screen.blit(text_surface, (10, y_offset + i * 20))

# Инициализация main для избежания использования функций в других файлах
if __name__ == "__main__":
    main()
```
